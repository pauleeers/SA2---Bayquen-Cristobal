Testing different distributions on the data in the Returns column

```python
dist_names = ['norm', 'lognorm', 'expon', 'gamma', 't', 'weibull_min', 'genextreme']
dist_results = []
for dist_name in dist_names:
    dist = getattr(stats, dist_name)
    params = dist.fit(df['Returns'])
    log_likelihood = np.sum(dist.logpdf(df['Returns'], *params))
    dist_results.append((dist_name, params, log_likelihood))
```

Testing different distributions again.

```python
dist_names = ['norm', 'lognorm', 'expon', 'gamma', 't', 'weibull_min', 'genextreme']
dist_results = []
for dist_name in dist_names:
    dist = getattr(stats, dist_name)
    params = dist.fit(df['Returns'])
    log_likelihood = np.sum(dist.logpdf(df['Returns'], *params))
    dist_results.append((dist_name, params, log_likelihood))
```

Showing results

```python
dist_results.sort(key=lambda x: x[2], reverse=True)
for dist_name, params, log_likelihood in dist_results:
    print(f'{dist_name.ljust(12)} {str(params).ljust(60)} {log_likelihood:.2f}')
```

Removing the distributions that have a negative infinite log_likelihood then running the test again

```python
dist_names = ['norm', 'lognorm', 'expon', 'gamma']
dist_results = []
for dist_name in dist_names:
    dist = getattr(stats, dist_name)
    params = dist.fit(df['Returns'])
    log_likelihood = np.sum(dist.logpdf(df['Returns'], *params))
    dist_results.append((dist_name, params, log_likelihood))
```

```python
dist_results.sort(key=lambda x: x[2], reverse=True)
for dist_name, params, log_likelihood in dist_results:
    print(f'{dist_name.ljust(12)} {str(params).ljust(60)} {log_likelihood:.2f}')
```

Since lognorm distribution has the highest log_likelihood, we'll test it with the Kolmogorov-Smirnov test.


```python
from scipy.stats import kstest, lognorm

# Select the distribution to test against
dist = lognorm

# Fit the distribution to the data
params = dist.fit(df['Returns'])

# Test the distribution against the data
ks_statistic, p_value = kstest(df['Returns'], dist.cdf, args=params)

# Print the results
print(f'KS statistic: {ks_statistic:.4f}')
print(f'P-value: {p_value:.4f}')
```
